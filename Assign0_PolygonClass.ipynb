```python
class Polygon:
    def __init__(self, sides):
        self.sides = sides
    
    def perimeter(self):
        return sum(self.sides)
    
    def num_sides(self):
        return len(self.sides)


class Rectangle(Polygon):
    def __init__(self, width, height):
        super().__init__([width, height, width, height])
    
    def area(self):
        return self.sides[0] * self.sides[1]
    
    def change_dimensions(self, new_width, new_height):
        print("Changed side parameters:")
        self.sides = [new_width, new_height, new_width, new_height]
    
    def report_dimensions(self):
        return (self.sides[0], self.sides[1])


class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)
    
    def area(self):
        return self.sides[0] ** 2
    
    def change_dimensions(self, new_side):
        self.sides = [new_side, new_side]
    
    def report_dimensions(self):
        return (self.sides[0], self.sides[0])


# define a function to display the menu and handle user input
def main_menu():
    while True:
        print("Please select an option:")
        print("1. Create a rectangle")
        print("2. Create a square")
        print("3. Quit")
        
        # get user input
        choice = input("> ")
        
        # create a rectangle
        if choice == "1":
            width = int(input("Enter the width: "))
            height = int(input("Enter the height: "))
            rect = Rectangle(width, height)
            print("Rectangle created")
            sub_menu(rect)
        
        # create a square
        elif choice == "2":
            side = int(input("Enter the side length: "))
            square = Square(side)
            print("Square created")
            sub_menu(square)
        
        # quit the program
        elif choice == "3":
            print("Exiting program...")
            break
        
        # invalid input
        else:
            print("Invalid choice. Please try again.")


# define a function to display the submenu and handle user input
def sub_menu(shape):
    while True:
        print("Please select an option:")
        print("1. Calculate area")
        print("2. Calculate perimeter")
        print("3. Change dimensions")
        print("4. Report dimensions")
        print("5. Return to main menu")
        
        # get user input
        choice = input("> ")
        
        # calculate area
        if choice == "1":
            print("The area is:", shape.area())
        
        # calculate perimeter
        elif choice == "2":
            print("The perimeter is:", shape.perimeter())
        
        # change dimensions
        elif choice == "3":
            new_dims = input("Enter the new dimensions (comma-separated): ")
            new_dims = [int(d.strip()) for d in new_dims.split(",")]
            if len(new_dims) == 2:
                shape.change_dimensions(*new_dims)
            else:
                print("Invalid input. Please try again.")
        
        # report dimensions
        elif choice == "4":
            print("The dimensions are:", shape.report_dimensions())
        
        # return to main menu
        elif choice == "5":
            print("Returning to main menu...")
            break
        
        # invalid input
        else:
            print("Invalid choice. Please try again.")


# run the main menu
main_menu()
```
